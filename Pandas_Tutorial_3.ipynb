# Different Ways Of Representing Data in Pandas 

## Through Dictionary

```python
import pandas as pd
Dict={'Album':['Perfect','Peaches','Wolves','Faded'],
      'Artist':['Ed Shereen','Justin Beiber','Selena Gomez','Alan Walker'],
      'Copies Sold(million)':[20,30,40,50]
        }
df=pd.DataFrame(Dict)
df
```

## Through Excel FIle

```python
df=pd.read_excel('Top 50.xlsx')
df.head()
```

## Through Tuples

```python
tup={('1/7/17',32,6,'Rain'),
     ('2/7/17',36,8,'Sunny'),
     ('3/7/17',30,10,'Rain')
     
    }
df=pd.DataFrame(tup,columns=['Day','Temp','WindSpeed','Event'])
df
```
