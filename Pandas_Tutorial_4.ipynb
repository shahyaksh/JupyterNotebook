# 1.Reading and writing data in csv file

### Skipping header rows from file using skiprows

```python
import pandas as pd
df=pd.read_csv('Top 50.csv',skiprows=2)
df
```

### Table with no header

```python
df=pd.read_csv('Top 50.csv',skiprows=2,header=None)
df
```

### Giving column names to table 

```python
df=pd.read_csv('Top 50.csv',skiprows=2,names=['Symbol','Open','High','Low','Last Traded Price','52 Week Low'])
df
```

### Replacing not availabie and n.a. value with NAN

```python
df=pd.read_csv('Top 50.csv',skiprows=2,na_values=['not available','n.a.',-1])
df
```

### Write the updated table in new csv file

```python
df.to_csv('new.csv',index=False)
df_new=pd.read_csv('new.csv')
df_new
```

# 2.Read From Excel File

```python
df=pd.read_excel('Top 50.xlsx')
df
```

### Convertor Function to change value in table

```python
def convert_Low_cell(cell):
    if cell=='not available':
        return None
    return cell
def convert_52WL_cell(cell):
    if cell=='n.a.':
        return None
    return cell
# convertors are important in Data cleaning or munjing
df=pd.read_excel('Top 50.xlsx',converters={
    'Low':convert_Low_cell,
    '52 Week Low':convert_52WL_cell
})
df
```

### Creating new excel file using to_excel 

```python
import pandas as pd
df=pd.read_excel('Top 50.xlsx',na_values=['not available','n.a.'])
df.to_excel('new_stock.xlsx',index=False)
df
```
